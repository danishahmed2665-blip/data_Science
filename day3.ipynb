**Methods of dictornary**

```python
d={"name":"Danish","age":18,"class":"12th"}

print(d.keys())
d.values()
print(d.values())

print(d.items())
d.update({"name":"Riyaz"})
print(d)
```

```python
#nested dict
d={"address":{"address1":{"city":"kukas city","district":"jaipur"},"address2":{"city":"gopalpura","district":"arya mains"}}}
print(d["address"]["address1"]["city"])
print(d["address"]["address1"]["district"])
print(d["address"]["address2"]["city"])
print(d["address"]["address2"]["district"])
```

```python
#nested list
l=[10,20,30,["hello","hello1","hello2",[True,False]]]
# print(l)
l1=l[3:4]
print(l1)
l2=l1[0][1:3]
print(l2)
```

```python
# map

def square(x):
  return x*x
l=[10,20,30]
l1= list(map(square,l))
print(l1)

# Map using lamda function
print(list(map(lambda x:x*x,l)))
```

```python
# Filter
l=["apple","banana","cat","dog"]
l1=list(filter(lambda x:x.endswith("a"),l))
print(l1)


# default function
def hello(x):
  return x.upper()
l1=list(map(hello,l))
print(l1)



# using lambda function
l2=list(map(lambda x:x.lower(),l1))
print(l2)
```

```python
# try except
try:
  num1=10
  num2=5
  print(num1/num2)

#Except

except:
  print("hello excepted")

# try except
try:
  num1=10
  num2=0
  print(num1/num2)
except:
  print("error occure")


# try except and finally
try:
  num1=10
  num2=5
except:
  print("error hae!")
finally:
  print(" mae to hamesha hi chalunga")
```
