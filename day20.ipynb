```python
#encoding

from sklearn.preprocessing import LabelEncoder
import pandas as pd

data = {
    'Name':['Ram','Kamal','Ajay','Neetu','Kavi','Sapu'],
    'Age':[25,None,32,20,21,22],
    'Salary':[5000,6000,None,7000,8000,9000],
    'Gender' : ['male','male','male','None','female','female']
}

df = pd.DataFrame(data)
print("original dataframe")
print(df)

# Label Encoder
le = LabelEncoder()
df_label = df.copy()
df_label['Gender'] = df['Gender'].fillna('unknown')
df_label['Gender_encoder'] = le.fit_transform(df_label['Gender'])
print("After label encoding:")
print(df_label)

```

**Handle missing values**

```python
import pandas as pd
import numpy as np
data = {
    "colors" : ['red','green','blue','orange','green','blue',np.nan]
}
df = pd.DataFrame(data)
#print(df)

#handle missing values
df.dropna(inplace=True)
print(df)
```

**Label Encoding**

```python
#print(df)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['colors_encoder'] = le.fit_transform(df['colors'])
print(df)
```

**One Hot Encoder**

```python

# One Hot Encoder
from sklearn.preprocessing import OneHotEncoder

print(df)

# 'colors_encoder' column delete
if 'colors_encoder' in df.columns:
    df.drop(['colors_encoder'], axis=1, inplace=True)

print(df)


# One Hot Encoder by the help of pandas
df_encoder = pd.get_dummies(df,columns=['colors'])
print(df_encoder)

# One Hot Encoder by the help of sklearn
encoder = OneHotEncoder()
encoder = encoder.fit_transform(df[['colors']])
print(encoded.array())
```
