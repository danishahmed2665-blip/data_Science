```python
import matplotlib.pyplot as plt  # visualization
x = [2010,2015,2020,2025]
y = [100,200,250,300]
plt.plot(x,y)  # Line Graph
plt.xlabel("years")  # x lable
plt.ylabel("sales")  # y lable
plt.title("sales report")  #graph title
plt.show()  #graph show
```

```python
#multiple line chart
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y1 = [100,200,260,290]
y2 = [150,185,195,300]

plt.plot(x,y1,label="jeans")
plt.plot(x,y2,label="shirt")
plt.legend()  # info of label
plt.show()



```

```python
#multiple line chart
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]   # Line 1
y2 = [1, 3, 5, 7, 9]    # Line 2
y3 = [2, 3, 4, 5, 6]    # Line 3

# Plot multiple lines
plt.plot(x, y1, label="Line 1", marker="o", linestyle="-", linewidth="4", markersize="14", color="b")
plt.plot(x, y2, label="Line 2", marker="s", linestyle=":", linewidth="8", markersize="20", color="m")
plt.plot(x, y3, label="Line 3", marker="^", linestyle="-.", linewidth="3", markersize="15", color="c")

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Chart Example")
plt.legend()

# Show chart
plt.show()
```

```python
#bar chart
import matplotlib.pyplot as plt
x = [2015,2020,2025,2030]
y = [100,150,200,290]

plt.figure(figsize=(6,2))
plt.bar(x,y)

plt.show()
```

```python
# multi bar chart
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y1 = [10,20,20,40]
y2 = [20,30,25,30]
# calculation -> width
w = 0.40
plt.bar(x - w/2,y1 , label="boys",width=w) # hide second
plt.bar(x + w/2,y2, label="girls",width=w) # show

plt.xlabel("groups")
plt.ylabel("number of students")
plt.title("Number of Students in Each group")
plt.legend()
plt.show()
```
