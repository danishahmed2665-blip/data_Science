```python
import pandas as pd
url="https://raw.githubusercontent.com/student-dheerajyadav/DataScience__45days/main/student-data.json"
df=pd.read_json(url)
# print(df)
#by default ascending
df.sort_values("english",inplace=True)
print(df)
print("--------------Descending----------------")
#descending
df.sort_values("name",ascending=False,inplace=True)
print(df)
print("-----------------------------------")
#by default ascending method 1.
print(df.sort_values(by=["maths","physics"],ascending=[False,True]))


print("---------------------------------")
#sortin in name
df["name"]=df["name"].str.upper()
df1=df.sort_values(by=["name"])
print(df1)
```

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    "name": ["A", "B", "C", "D", "E", "F"],
    "marks": [50, 60, 70, 80, 90, 100]
})

# ✅ Row delete (index 6 से 14 तक)
df = df.drop(df.index[6:15])

# ✅ नया column 'doj' add करना
df["doj"] = ["2025-01-01","2025-02-02","2025-03-03","2025-04-04","2025-05-05","2025-06-06"]

print(df)

# ✅ String → Date format में convert करना
df["doj"] = pd.to_datetime(df["doj"])
print(df)

# ✅ Date operations
print(df["doj"].dt.year)       # Year निकालना
print(df["doj"].dt.month)      # Month निकालना
print(df["doj"].dt.day)        # Day निकालना
print(df["doj"].dt.day_name()) # Day name निकालना

# ✅ 20 days add करना
print(df["doj"] + pd.to_timedelta(20, unit="D"))
```
