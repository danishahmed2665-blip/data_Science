```python
import pandas as pd
url = "https://raw.githubusercontent.com/rajendra0968jangid/Ds-Arya/main/student-data.json"
df = pd.read_json(url)

#by default ascending
df.sort_values("english")

# descending
df.sort_values("english", ascending=False)

#by default ascending method
df.sort_values(by=["english","maths"],ascending=[False,True])

#
df.sort_values("name")

#a to z
df["name"] = df["name"].str.lower()
df.sort_values("name")

#add column total
df["total"] = df['english']+df['maths']+df['physics']
df

#add row
df.loc[14] = ["danish",'Male',50,80,80,210]
df

#delete column
df.drop("total",axis=1)


#delete row
df.drop(14,axis=0)


df.drop([6,7,8,9,10,11,12,13,14],inplace=True)
df

df["doj"] = ['2025-01-01','2025-02-02','2025-03-03','2025-04-04','2025-05-05','2025-06-06']
df
```
